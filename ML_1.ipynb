```python
def comp_cost(x , y , w , b) :
  m = len(x)
  y_pred = w*x+b
  cost = (1/(2*m)*np.sum((y_pred - y)**2))
  return cost

def comp_grad(x , y , w , b) :
  m = len(x)
  y_pred = w*x+b
  dw = (1/m)*np.sum(((y_pred-y))*x)
  db = (1/m)*np.sum(((y_pred-y)))
  return dw , db

def grad_desc(x , y , w , b , alpha , iter) :
  for i in range(iter) :
    dw , db = comp_grad(x , y , w , b)
    w = w - alpha*dw
    b = b - alpha*db
  return w , b
```

```python
w = 0
b = 0
alp = 0.001
iterations = 25000
w, b = grad_desc(x_train, y_train, w, b, alp, iterations)
comp_cost(x_train , y_train , w , b)
pred = float(input("Enter CGPA (out of 10): " ))
if(pred>10 or pred<0) :
  print("Invalid CGPA.")
  exit()
else :
  sal = w*pred+b
print(f"Salary is: {sal} lpa")

```
